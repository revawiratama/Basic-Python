```python
# Let's make a visualization of more than one log from a new data set

import lasio
las = lasio.read('./data/15_9-13.las.txt')
data2 = las.df()
data2
```

```python
data2.isna().sum()/data2.shape[0]
```

```python
# let's make the index which is our depth to be a column of the dataframe for better log plotting use

data2['DEPTH'] = data2.index
data2.head()
```

```python
# we will be making plots of the gamma ray, sp, neutron porosity and bulk density on the same figsize for visualization
# so we will have to employ a special feature for that, the subplot method

top = data2['DEPTH'].iloc[0]
base = data2['DEPTH'].iloc[data2.shape[0] - 1]

f, ax = plt.subplots(nrows=1, ncols=4, figsize=(12,10))  # 4 columns on a single figure row

#iterating through the created figures to set individual axes 
for i in range(len(ax)):
    ax[i].set_ylim(top, base)
    ax[i].invert_yaxis()
    ax[i].grid()
    ax[i].locator_params(axis='x', nbins=4)

    #plotting the actual data points on the created log columns
    ax[0].plot(data2.GR, data2.DEPTH)
    ax[1].plot(data2.NPHI, data2.DEPTH)
    ax[2].plot(data2.RHOB, data2.DEPTH)
    ax[3].plot(data2.SP, data2.DEPTH)
```

```python
# We could make the visualization more appealing by adding some more informations and editing the axes

f, ax = plt.subplots(nrows=1, ncols=4, figsize=(12,10))  # 4 columns on a single figure row

#iterating through the created figures to set individual axes 
for i in range(len(ax)):
    ax[i].set_ylim(top, base)
    ax[i].invert_yaxis()
    ax[i].grid()
    ax[i].locator_params(axis='x', nbins=4)

    #plotting the actual data points on the created log columns
    ax[0].plot(data2.GR, data2.DEPTH, color='black')
    ax[1].plot(data2.NPHI, data2.DEPTH, color='green')
    ax[2].plot(data2.RHOB, data2.DEPTH, c='blue')
    ax[3].plot(data2.SP, data2.DEPTH, c='red')
    
    # adding axes information and adjusting axes
    ax[0].set_xlabel('GR')
    ax[0].set_xlim(0, 150)
    ax[1].set_xlabel('NPHI')
    ax[2].set_xlabel('RHOB')
    ax[3].set_xlabel('SP')
    
    f.suptitle('Log plots of well 15/9-13', fontsize=14, y=0.92)
```
