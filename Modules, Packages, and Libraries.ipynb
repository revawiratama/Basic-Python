# MODULES, PACKAGES, LIBRARIES

We’ve seen how functions can be written and used in a python program. What if we need these functions in a different program. Do we have to rewrite or redefine the functions in our new program? Well, we don’t have to. This brings us to the concept of Modules.

**Modules** are python files which are used in another program. They could be functions written in a different file, or a script used to solve a particular task and then imported into the main python file. This is very useful as we can write functions to be reused in many other programs. This also brings us to the concept of open source. 
Open source refers to software which is released to the public under a license (usually free to all). This includes the source, and applications. Open source also extends to softwares in which the development process is open to contributions from the general public. Like we mentioned at the start of the course, Python is open source. In essence, the whole concept of open source is that you can use other people’s codes to solve problems without having to copy or rewrite the codes. The codes are usually in the form of modules (python files), packages and libraries. A **package** is a collection of similar python files used for a specific task while a library is a collection of similar packages and **modules** designed for a particular type of task.

### How modules, packages and libraries can be used.

To use modules which you've created on your local system in a different program, the modules could be copied to the same folder in which the main program is. The **import** keyword is used to import modules into the python program which need them.

#### Example

We are going to create a python module which will be used for arithmetic operations in a different python program. This module will have functions for basic arithmetic operations; addition, subtraction, multiplication and division. The main python file which needs these functions/this module is this current notebook while the module will be named **arithmetic**.

```python
import arithmetic
```

Importing a module which does not exist in our working directory throws back a **ModuleNotFoundError**

So how do we access our functions after importing this module. We simply use the dot method which was used when dealing with data structures and their methods. In Jupyter Notebook, after typing out the module and adding a dot, the **tab** key can be used to display the available functions -- methods in this case -- that are available for that module

```python
arithmetic.add(12, 1000)
```
